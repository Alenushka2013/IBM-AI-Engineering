# Task A

Use the Keras library to build a neural network with the following:

- One hidden layer of 10 nodes, and a ReLU activation function

- Use the adam optimizer and the mean squared error  as the loss function.

1. Randomly split the data into a training and test sets by holding 30% of the data for testing. You can use the

helper function from Scikit-learn.

2. Train the model on the training data using 50 epochs.

3. Evaluate the model on the test data and compute the mean squared error between the predicted concrete strength and the actual concrete strength. You can use the mean_squared_error function from Scikit-learn.

4. Repeat steps 1 - 3, 50 times, i.e., create a list of 50 mean squared errors.

5. Report the mean and the standard deviation of the mean squared errors.

```python
import pandas as pd
import numpy as np
import keras

import warnings
warnings.simplefilter('ignore', FutureWarning)
```

```python
filepath='https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv'
concrete_data = pd.read_csv(filepath)

concrete_data.head()
```

```python
concrete_data_columns = concrete_data.columns
```

```python
predictors = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']] # all columns except Strength
target = concrete_data['Strength'] # Strength column
```

```python
predictors.head()
```

```python
target.head()
```

```python
n_cols = predictors.shape[1] # number of predictors
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Input

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
# define regression model
def regression_model():
    # create model
    model = Sequential()
    model.add(Input(shape=(n_cols,)))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1))

    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
# build the model
model = regression_model()

mean_squared_errors = []
for _ in range(50):
  X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size = 0.3)
  model.fit(X_train, y_train, epochs=50, verbose=2)
  predictions = model.predict(X_test)
  mean_squared_errors.append(mean_squared_error(y_test, predictions))
```

```python
print(np.array(mean_squared_errors).mean(), np.array(mean_squared_errors).std())
```

# Task B

B. Normalize the data (5 marks)

Repeat Part A but use a normalized version of the data. Recall that one way to normalize the data is by subtracting the mean from the individual predictors and dividing by the standard deviation.

How does the mean of the mean squared errors compare to that from Step A?

```python
model_b = regression_model()

mean_squared_errors_norm = []
for _ in range(50):
  X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size = 0.3)
  X_train_norm = (X_train - X_train.mean()) / X_train.std()
  X_test_norm = (X_test - X_test.mean()) / X_test.std()
  model_b.fit(X_train_norm, y_train, epochs=50, verbose=2)
  predictions_norm = model_b.predict(X_test_norm)
  mean_squared_errors_norm.append(mean_squared_error(y_test, predictions_norm))
```

```python
print(np.array(mean_squared_errors_norm).mean(), np.array(mean_squared_errors_norm).std())
```

*How does the mean of the mean squared errors compare to that from Step A?*


In step B the mean of the mean squared errors is less than in step A.

# Task D

D. Increase the number of hidden layers (5 marks)

Repeat part B but use a neural network with the following instead:

- Three hidden layers, each of 10 nodes and ReLU activation function.

How does the mean of the mean squared errors compare to that from Step B?

```python
# define regression model
def regression_model_d():
    # create model
    model = Sequential()
    model.add(Input(shape=(n_cols,)))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1))

    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
model_d = regression_model_d()
mean_squared_errors_norm_d = []
for _ in range(50):
  X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size = 0.3)
  X_train_norm = (X_train - X_train.mean()) / X_train.std()
  X_test_norm = (X_test - X_test.mean()) / X_test.std()
  model_d.fit(X_train_norm, y_train, epochs=50, verbose=2)
  predictions_norm = model_d.predict(X_test_norm)
  mean_squared_errors_norm_d.append(mean_squared_error(y_test, predictions_norm))
```

```python
print(np.array(mean_squared_errors_norm_d).mean(), np.array(mean_squared_errors_norm_d).std())
```
